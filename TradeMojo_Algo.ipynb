```python
import gspread
import pandas as pd
from oauth2client.service_account import ServiceAccountCredentials
from nsetools import Nse


xlrow = 1
xlcol = 1

# defining the scope of the application
scope_app =['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive'] 

#credentials to the account
cred = ServiceAccountCredentials.from_json_keyfile_name('E:/Abhyasana/TradeMojo/trademojoalgo-17e2df8912fd.json',scope_app) 

# authorize the clientsheet 
client = gspread.authorize(cred)
# get the sample of the Spreadsheet
sheet = client.open('TradeMojo_Algo')

# get the first sheet of the Spreadsheet
sheet_instance = sheet.get_worksheet(0)

nse=Nse()
Gainers=nse.get_top_gainers()
Gainers=pd.DataFrame(Gainers)
Losers=nse.get_top_losers()
Losers=pd.DataFrame(Losers)
BUY=[x for x in Gainers['symbol']]
LTP=[x for x in Gainers['ltp']]
SELL=[x for x in Losers['symbol']]
for x in range(len(BUY)):
    #print(BUY[x])
    sheet_instance.update_cell(xlrow+1, xlcol, BUY[x])
    sheet_instance.update_cell(xlrow+1, xlcol+7, LTP[x])
    #sheet_instance.update_cell(xlrow+1, xlcol+8, nse.get_stock_codes(BUY[x]))
       
    
    xlrow = xlrow +1
    

# (Column, Row)
#sheet_instance.update_cell(2, 1, "BIDEN VOTES")

# get all the records of the data
records = sheet_instance.get_all_values()

# view the data
records

# convert the json to dataframe
records_df = pd.DataFrame.from_dict(records)

# view the top records
records_df.head()
```
